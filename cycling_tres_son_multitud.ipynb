```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
```

```python
user_agent = {'User-agent': 'Mozilla/5.0'}
lista_tours = []
lista_clasicas = []

def tours_lista(race_url, start_year):
    base_url = 'https://www.procyclingstats.com/race/' + str(race_url) + '/'
    print(str(race_url).upper(),'inventorio...')
    for year in range(int(start_year),int(2023)):
        for stage in range(1,22):
            url = base_url + str(year) + '/stage-' + str(stage) + "/result/result"
            lista_tours.append(url)
```

```python
def clasicas_lista(race_url, start_year):
    base_url = 'https://www.procyclingstats.com/race/' + str(race_url) + '/'
    print(str(race_url).upper(),'starting...')
    for year in range(int(start_year),int(2023)):
        url = base_url + str(year) + "/result"
        lista_clasicas.append(url)
```

```python
# hay que guardar el formato de la pagina de procyclingstats para el nombre de la carrera
tours = ["paris-nice", "volta-a-catalunya", "itzulia-basque-country", "tour-de-romandie", "giro-d-italia",
        "dauphine", "tour-de-suisse", "tour-de-france", "tour-de-pologne", "vuelta-a-espana", "benelux-tour"]
clasicas = ["omloop-het-nieuwsblad", "strade-bianche", "milano-sanremo", "oxyclean-classic-brugge-de-panne",
           "e3-harelbeke", "gent-wevelgem", "dwars-door-vlaanderen", "ronde-van-vlaanderen", "amstel-gold-race",
           "paris-roubaix", "la-fleche-wallone", "liege-bastogne-liege", "Eschborn-Frankfurt", 
           "san-sebastian", "cyclassics-hamburg", "bretagne-classic", "il-lombardia"]
```

```python
for race in tours:
    tours_lista(race,1990)
    print("lista_tours creada")
    
for race in clasicas:
    clasicas_lista(race,1990) 
    print("lista_clasicas creada")
```

```python
def probando(lista_races):
    for i in lista_races:
        response = requests.get(i, headers=user_agent)
        if response.status_code == 200:
            try:
                page = response.text
                soup = BeautifulSoup(page, 'lxml')
                ooo = soup.find("table")
                table_body = ooo.find('tbody')
                data = []
                rows = table_body.find_all('tr')[0:3]
                for row in rows:
                    cols = row.find_all('td')
                    cols = [ele.text.strip() for ele in cols]
                    uuu = [w for w in cols if len(w) >= 3]
                    data.append(uuu)
                df = pd.DataFrame(data)
                eee = df.stack().value_counts()
                if eee[0] == 3:
                    print(i, eee.index[0])
                else:
                    print(i, "..." )
            except:
                pass
        else:
            print('unsuccessful request!')
            print('status code:',response.status_code)
            break
```

```python
probando(lista_clasicas)
# comprueba si los tres primeros son del mismo equipo en todas las carreras especificadas, una por una
# si la condicion se cumple imprime el nombre del equipo, en caso contrario imprime "..."
```

```python
probando(lista_tours)
```
